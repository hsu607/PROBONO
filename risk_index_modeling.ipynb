```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data_all.csv", )
df['IP_PM2.5'] = ""
df['IP_PM10'] = ""
df['IP_O3'] = ""
df['IP_NO2'] = ""
df['IP_CO'] = ""
df['IP_SO2'] = ""
for i in range(0,730):
  if df['PM2.5'][i] <= 15:
    BP_L = 0
    BP_H = 15
    I_L = 0
    I_H = 50
  elif df['PM2.5'][i] <= 35:
    BP_L = 16
    BP_H = 35
    I_L = 51
    I_H = 100
  elif df['PM2.5'][i] <= 75:
    BP_L = 36
    BP_H = 75
    I_L = 101
    I_H = 250
  elif df['PM2.5'][i] <= 500:
    BP_L = 76
    BP_H = 500
    I_L = 251
    I_H = 500
  else:
    BP_L = 76
    BP_H = 500
    I_L = 251
    I_H = 500
  df['IP_PM2.5'][i] = (I_H -I_L)/(BP_H - BP_L) * (df['PM2.5'][i] - BP_L) + I_L

df['IP_PM10'] = ""
for i in range(0,730):
  if df['PM10'][i] <= 30:
    BP_L = 0
    BP_H = 30
    I_L = 0
    I_H = 50
  elif df['PM10'][i] <= 80:
    BP_L = 31
    BP_H = 80
    I_L = 51
    I_H = 100
  elif df['PM10'][i] <= 250:
    BP_L =81
    BP_H = 150
    I_L = 101
    I_H = 250
  elif df['PM10'][i] <= 600:
    BP_L = 151
    BP_H = 600
    I_L = 251
    I_H = 500
  else:
    BP_L = 151
    BP_H = 600
    I_L = 251
    I_H = 500
  df['IP_PM10'][i] = (I_H -I_L)/(BP_H - BP_L) * (df['PM10'][i] - BP_L) + I_L

df['IP_O3'] = ""
for i in range(0,730):
  if df['O3'][i] <= 0.03:
    BP_L = 0
    BP_H = 0.03
    I_L = 0
    I_H = 50
  elif df['O3'][i] <= 0.09:
    BP_L = 0.031
    BP_H = 0.09
    I_L = 51
    I_H = 100
  elif df['O3'][i] <= 0.15:
    BP_L = 0.091
    BP_H = 0.15
    I_L = 101
    I_H = 250
  elif df['O3'][i] <= 0.6:
    BP_L = 0.151
    BP_H = 0.6
    I_L = 251
    I_H = 500
  else:
    BP_L = 0.151
    BP_H = 0.6
    I_L = 251
    I_H = 500
  df['IP_O3'][i] = (I_H -I_L)/(BP_H - BP_L) * (df['O3'][i] - BP_L) + I_L

df['IP_NO2'] = ""
for i in range(0,730): 
  if df['NO2'][i] <= 0.03:
    BP_L = 0
    BP_H = 0.03
    I_L = 0
    I_H = 50
  elif df['NO2'][i] <= 0.06:
    BP_L = 0.031
    BP_H = 0.06
    I_L = 51
    I_H = 100
  elif df['NO2'][i] <= 0.2:
    BP_L = 0.061
    BP_H = 0.2
    I_L = 101
    I_H = 250
  elif df['NO2'][i] <= 2:
    BP_L = 0.201
    BP_H = 2
    I_L = 251
    I_H = 500
  else:
    BP_L = 0.201
    BP_H = 2
    I_L = 251
    I_H = 500

  df['IP_NO2'][i] = (I_H -I_L)/(BP_H - BP_L) * (df['NO2'][i] - BP_L) + I_L

df['IP_CO'] = ""
for i in range(0,730):         
  if df['CO'][i] <= 2:
    BP_L = 0
    BP_H = 2
    I_L = 0
    I_H = 50
  elif df['CO'][i] <= 9:
    BP_L = 2.1
    BP_H = 9
    I_L = 51
    I_H = 100
  elif df['CO'][i] <= 15:
    BP_L = 9.1
    BP_H = 15
    I_L = 101
    I_H = 250
  elif df['CO'][i] <= 50:
    BP_L = 15.1
    BP_H = 50
    I_L = 251
    I_H = 500
  else:
    BP_L = 15.1
    BP_H = 50
    I_L = 251
    I_H = 500  

  df['IP_CO'][i] = (I_H -I_L)/(BP_H - BP_L) * (df['CO'][i] - BP_L) + I_L

df['IP_SO2'] = ""
for i in range(0,730):         
  if df['SO2'][i] <= 0.02:
    BP_L = 0
    BP_H = 0.02
    I_L = 0
    I_H = 50
  elif df['SO2'][i] <= 0.05:
    BP_L = 0.021
    BP_H = 0.05
    I_L = 51
    I_H = 100
  elif df['SO2'][i] <= 0.15:
    BP_L = 0.051
    BP_H = 0.15
    I_L = 101
    I_H = 250
  elif df['SO2'][i] <= 1:
    BP_L = 0.151
    BP_H = 1
    I_L = 251
    I_H = 500
  else:
    BP_L = 0.151
    BP_H = 1
    I_L = 251
    I_H = 500

  df['IP_SO2'][i] = (I_H -I_L)/(BP_H - BP_L) * (df['SO2'][i] - BP_L) + I_L

df['IP'] = df[['IP_PM2.5','IP_PM10','IP_O3','IP_NO2','IP_CO','IP_SO2']].max(axis = 1)
df.tail(3)
```

```python
# import package
import random

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/IP.csv", )

X = df.drop(columns = ['Unnamed: 0','harm', 'death', 'treatment', 'date','IP_PM2.5','IP_PM10','IP_O3','IP_NO2','IP_CO','IP_SO2','PM2.5','PM10','O3','NO2','CO','SO2'])
Y = df[['death']]
# train/test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y,

                                                   shuffle = True,

                                                   test_size = 0.2,

                                                   random_state=1)

# feature scaling
from sklearn import preprocessing

# normalization
minmax_scaler = preprocessing.MinMaxScaler()
norm_fit = minmax_scaler.fit(X_train)
X_train_norm = norm_fit.transform(X_train)
X_test_norm = norm_fit.transform(X_test)

# standardization
standard_scaler = preprocessing.StandardScaler()
stan_fit = standard_scaler.fit(X_train)
X_train_stan = stan_fit.transform(X_train)
X_test_stan = stan_fit.transform(X_test)

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

def build_model(num_input=1):
    model = Sequential()
    
    # hidden layer
    model.add(Dense(128, activation='relu', input_dim=num_input))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(16, activation='relu'))
    
    # output layer
    model.add(Dense(1, activation='linear'))
    
    # model compile
    model.compile(optimizer='adam', loss='mse', metrics=['mse'])
    
    return model

model = build_model(num_input=8)   # feature의 개수 = 8
model.summary()
```

```python
EPOCHS = 200

history = model.fit(X_train_stan, y_train, batch_size = 32,
                    epochs=EPOCHS, validation_split = 0.2, verbose=0) 

hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()

# loss function
def plot_loss_curve(total_epoch = 10, start = 1):
    
    # packge for visualization
    import matplotlib.pyplot as plt
    import seaborn as sns
    sns.set_theme(color_codes = True)
    
    plt.figure
    plt.plot(figsize = (15,5))
    plt.plot(range(start, total_epoch +1), history.history['loss'][start-1:total_epoch], 
             label = "Train")
    plt.plot(range(start, total_epoch +1), history.history['val_loss'][start-1:total_epoch], 
             label = "Validation")
    plt.xlabel("Epochs")
    plt.ylabel("loss")
    plt.legend()
    plt.show()

# mse function
def plot_mse_curve(total_epoch = 10, start = 1):
    
    # packge for visualization
    import matplotlib.pyplot as plt
    import seaborn as sns
    sns.set_theme(color_codes = True)
    
    plt.figure
    plt.plot(figsize = (15,5))
    plt.plot(range(start, total_epoch +1), history.history['mse'][start-1:total_epoch], 
             label = "Train")
    plt.plot(range(start, total_epoch +1), history.history['val_mse'][start-1:total_epoch], 
             label = "Validation")
    plt.xlabel("Epochs")
    plt.ylabel("MSE")
    plt.legend()
    plt.show()



plot_loss_curve(total_epoch=40, start=1)

plot_mse_curve(total_epoch=40, start=1)

res = model.evaluate(X_test_norm, y_test, verbose=2)  # loss = res[0], mse = res[1]

print(f"테스트 세트의 MSE: {round(res[1],3)*1000}")
```
